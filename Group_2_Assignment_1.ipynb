# **Logging Company**

## **Idea:**
One of our ideas was to have our database be for a logging company to help keep track of important data. We think this idea is very unique in the sense that a logging company is not  something that most people would think about for creating a database about and it seems fairly abstract but still relevant. We would be borrowing ideas from already existing logging companies to find out most of our data tables. A typical business day would most likely consist of keeping records on the amount of employees working, the areas that are being cut down, as well as the number of trees being cut on an average work day. There would, also, be employees surveing new land or current land for logging to be conducted. 

## **Tables**

**1.**  **Number of Trees Cut** : keeping track of the amount of trees that are cut down in an area at one time or over the course of a project to avoid deforestation. Keeping track of daily amounts of timber that has been cut to help maintain some form of quota.


**2.** **Trees Processed** :  Keep track of the amount of trees that are being processed or cut down into timber to be sold to companies over an amount of time.

**3.**  **Amount of machinery** : keeping track of how much heavy machinery is being used at a given time to help with workload. Examples being transportation trucks, as well as anything that would help workers cut down trees.

**4.**   **Region** : Area/location that the timber company is currently working at in terms of cutting down trees.


**5.**   **Man Hours** : keeping a record of the amount of employees working at a given time, and also the amount of hours that said employees are working. The man hours table will be the one containing our dynamic data that will change based on number of employees working.


## **Entities and Attributes**

* **Number of Trees Cut**: Entity - Tree cut down. Attributes - Type of tree, height of tree, Weight of tree. Constraints - No trees under a certain height.

* **Trees Processed**: Entity - Tree being processed. Attributes -Type of tree, height of tree, Weight of tree, Amount of timber gathered 

* **Region**: Entity - Local forest. Attributes - Size of forest, Wildlife in the area. Constraints: No areas with endangered animals or protected forests.

* **Amount of Machinery**: Entity - Type of machine. Attributes - Run time, operator experience, maintenance of the machine

* **Man Hours**: Entity - Employee. Attributes - Work hours, position, name

## **Goal**

* The goal for this project is to be able to better schedule employees and identify any under or over used machinery. This will be accomplished by using "Trees Processed" & "Man Hours" tables.

#Penn State Football Stats

## Idea
We had the idea of creating a database used for looking up statistics for the Penn State Football team. In this database, we would include the players on the roster along with their number and what position they play. We could also include where they went to and their hometown. Some of the other ideas were to include passing yards for the Quarterbacks, rushing yards for the running backs, and receiving yards for the receivers. For the player's stats, we would include how many games they have played and how many games they have started. We would need to use other sources to gather information such as ESPN or from the Penn State football team's website. The tables that we would be using are as follows: players, offense, defense, special teams, and coaches. In each of the tables we would be able to include stats that would be relevant to know about the specific table. Things like the players or coaches hometown would not be dynamic data but numerical statistics would be. Rushing, receiving, and passing yards would all change every game which would make them all dynamic data. Some of the constraints are needed because things like coaches' names or player names cant be an integer. Within these constraints, the name can't be longer than the specified length. The goal for this project is to be able to pick a better starting line fo rthe next game.
