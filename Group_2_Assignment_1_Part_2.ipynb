**Idea 2: Restaurant**

We had the idea of creating a database used for collecting valuable data for a restaurant.  While this idea is not new, it is unique in the sense that this data could be collected in one place rather than a restaurant figuring out these numbers through experience.  We would be borrowing ideas from an existing restaurant (not yet decided).   A typical business day consists of several dynamic parts. First we have the logistics, the chefs and managers charged with ensuring there will be enough product to fit the expected demand. This position, manager mainly, may also include purchasing a variety of other items. Then we have the consumer interactions. These actions are pretty standard, with actions such as din-in or take-out. It also includes what items are purchases the costumer makes. Then we have the production group, the cooks. These are the ones who do the food prep, cooking and dishes.  The collection of this data could eliminate confusion on when to schedule people to work, eliminate any possibility of running out of ingredients, as well as test out a new dish by being able to see how many people order it. More specifically we are looking to understand and evaluate what customers are buy in order to cater the menu to that area of cuisine. We do need to put in some constraints to help keep erroneous data out of the database. Mainly we have a no negative value for customers, price, quantity and hours worked.
