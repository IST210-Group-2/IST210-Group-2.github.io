**Question 1:**
What is the average salary of full time chefs?

```
cursor.execute('''select avg(Salary) from Employee where Status = "full" and Position = 'chef' ''')
for row in cursor.fetchall():
  print(row)
```

The first question, “What is the average salary of full time chefs?” is important because we want to make sure that on average, full time chefs are being paid in keeping with industry or domain averages. This ensures that the restaurant as a business remains competitive and does not lose chefs to better salaries elsewhere. Our average is slightly below the industry average, indicating to us we should increase chef wages.

**Question 2:**
How many recent events were birthdays?

```
cursor.execute('''select count(eventID) from Events where type = "birthday" ''')
for row in cursor.fetchall():
  print(row)
```

The second question asked how many recent events were birthdays.  This question is important for a restaurant to be able to keep track of how many events they had scheduled were birthdays.  This information could allow them to gain knowledge on their most popular events and perhaps where to focus their energy.  Our results matched up with our expectations due to the fact that as a restaurant it is expected to have more birthdays scheduled than weddings.   

**Question 3:**
Which loyalty program member has the most points and what is their discount?

```
cursor.execute('''select Fname, Lname, max(Loyalty_Points), discount from Loyalty ''')
for row in cursor.fetchall():
  print(row)
```

The third question asked which loyalty member had the most points and what was this individual’s discount.  This question is important for a restaurant to be able to gauge whether the amount of loyalty points an individual has results in a fair discount amount.  Our results showed that the individual with the most loyalty points had a discount of 5% and matched up with our expectations.
