When we were creating our ER diagram for a restaurant, we beleive we have included all of the nessescities that would be required to operate daily while keeping track of everything. The entities we have created in our diagram are Income, Expenses, Employee, Dishes, Menu, Events, Menu Sales, and Loyalty. Thinking about everything that is needed in a restaurant and what can help make a restaurant sucessful, these entities cover all bases and give us enough to work with to make our diagram seem full. Our Income entity has the primary key of IncomeID and the Forign keys from events and sales. The primary key for the Events entity is simply eventID with the attributes being some example of possible events. For example, birthdays, banquets, and graduations are the three we included into out ER diagram. The Menu Sales entity comtains the primary key saleID with multiple subkeys that are, entreeID, drinkID, and appetizerID. The attribute that is total_cost and a forign key being the customers. Our restaurant also has a loyalty program which has a primary key for customerID to take all of the customers personal information including email and phone number, to keep track of there loyalty points they have. The Menu is also its own entity with a PK of menuID and it has multiple attributes regarding the customers choice of meal and the price of said meal with a forign key then linking back to the menu sales. This then leads to our entity for Dishes with the primary key of DishID, this category has some interesting attributes for ingredients, cook time, cost, and then has a forign key linking back to the Menu entity. After the Dishes entity we have an that linked to an entity calulating Expenses with the PK of PurchaseID and a forign key for dishes and employees which will be where we incur most of our losses. The attribues for expenses also include ingredients, as well as, cutlery, and advertisment. And our final entity is Employee which keeps track of all our employees information included hours worked, salary, position and status.
