**Description**: In our groups ER diagram, we have seven different entities to represent our diagram. Our entities are Purchases, Fish, Aquarium, Pets, Enclosure, Loyalty Program, and Expence Report. Based on the description of the pet shop that we are running, these catagories seemed like good choices to base the diagram on while giving us enough detail to flesh out the ER diagram. The Purchases entity with our primary key being PurchaseID, contains attributes for fish purchases, pet purchases, food purchases, and customerLoyaltyID as a foreign key to check to see if the customer purchaseing an item is a loyalty memeber. Our Fish entity with the primary key being the fishID, has a secondary key that is the type of fish being purchased, and a forign key being the aquarium the fish is held in. The other two attributes that we have are the price of the fish and the food needed to feed it. The connectivity and cardinality between purchases and fish is 1 purchase to the possibility of many fish. The Pets entity is set almost identically to the fish entity with the only main differences being that pets are held in an enclosure and fish and held in an aquarium. The last entity that is linked to the purchases one is the Loyalty Program. The primary key for this is the customerLoyaltyID and it has multiple secondary keys, Customer First Name, Customer Last Name, and Customer Email. The other four attributes for the Loyalty Program are the number of fish purchases, pet purchases, food purchases, and then the total amount of loyalty points that the customer has accumulated through all the purchases.The Aquarium entity and the enclosure entity are almost the same, the only different being that they deal with there respective category of pet or fish. The primary key for Aquarium is aquariumID and the primary key for Enclosure is enclosureID. There attributes are location and occupancy for both of them. They both share a 1 to many relationship with the pets or fish entitys. The last entity is the expense report with the primary key being the expenseID. The attributes for this are cost of fish food and cost of pet food. This relationship is completly reliant upon the number of pets and fish that are in our store.
